```python
import time
import datetime
import numpy as np
```

```python
start_time = time.time()
```

```python
def read_file():
    file = "input.txt"
    #file = "testi.txt"
    result = [0, 0]
    with open(file, "r") as inputfile:
         data = np.array([i for i in inputfile.read().splitlines()])
    rules = np.array([['0', '0000'],
         ['1', '0001'],
         ['2', '0010'],
         ['3', '0011'],
         ['4', '0100'],
         ['5', '0101'],
         ['6', '0110'],
         ['7', '0111'],
         ['8', '1000'],
         ['9', '1001'],
         ['A', '1010'],
         ['B', '1011'],
         ['C', '1100'],
         ['D', '1101'],
         ['E', '1110'],
         ['F', '1111']])
    result[0] = rules
    result[1] = data
    return result
```

```python
def run_task(task):
    t0 = time.time()
    data = read_file()
    if task == 1:
        first_star(data)
    else:
        second_star(data)
    t1 = time.time()
    # print(data)
    print("tehtävän ajoon aikaa meni:", round(t1-t0,4), "s")
    print('tehtävän suorittamiseen meni:', datetime.timedelta(seconds=round(t1-start_time)), "s")
```

```python
def debug_transmission(binary, version_sum):
    # print(binary)
    ver = int(binary[:3], 2)
    version_sum += ver
    binary = binary[3:]
    ID = int(binary[:3], 2)
    binary = binary[3:]
    print('version',ver, '\nID',ID)
    if ID != 4:
        ID_L = int(binary[0])
        binary = binary[1:]
        if ID_L:
            amount = binary[:11]
            binary = binary[11:]
            # print(ID_L, 'Amount', int(amount, 2))
            rounds = 0
            while rounds < int(amount, 2):
                binary, version_sum = debug_transmission(binary, version_sum)
                rounds += 1
        else:
            lenght_sub = binary[:15]
            binary = binary[15:]
            # print(ID_L,'lenght', int(lenght_sub,2))
            binary_temp = binary[::]
            while len(binary_temp) > 0:
                binary_temp, version_sum = debug_transmission(binary_temp[:int(lenght_sub, 2)], version_sum)
            binary = binary[int(lenght_sub, 2):]
    else:
        message = ''
        # print('4',binary)
        while len(binary) >= 5:
            message += binary[1:5]
            if not int(binary[0]):
                binary = binary[5:]
                break
            binary = binary[5:]
        print('Message',int(message, 2))
    # print('Lopulta', binary, version_sum)
    return binary, version_sum
    
```

```python
def first_star(data):
    # print(data)
    rules = list(zip(*data[0]))
    transmission = data[1]
    All_ver_sum = 0
    for command in transmission:
        binary = ''
        # print(command)
        for char in command:
            binary += rules[1][rules[0].index(char)]
        jaljella, ver_sum = debug_transmission(binary, 0)
        All_ver_sum += ver_sum
    print('yhteensä',All_ver_sum)
    return 0
```

```python
def debug_transmission2(binary, version_sum, sum_all):
    # print(binary)
    ver = int(binary[:3], 2)
    binary = binary[3:]
    ID = int(binary[:3], 2)
    binary = binary[3:]
    # print('\nID',ID)
    if ID != 4:
        ID_L = int(binary[0])
        binary = binary[1:]
        # sum_all = []
        if ID_L:
            amount = binary[:11]
            binary = binary[11:]
            # print(ID_L, 'Amount', int(amount, 2))
            rounds = 0
            temp_all = []
            while rounds < int(amount, 2):
                binary, version_sum,  temp_all= debug_transmission2(binary, version_sum,  temp_all)
                rounds += 1
            # print(temp_all)
            sum_all.append(do_the_function(temp_all, ID))
        else:
            lenght_sub = binary[:15]
            binary = binary[15:]
            # print(ID_L,'lenght', int(lenght_sub,2))
            binary_temp = binary[::]
            temp_all = []
            while len(binary_temp) > 0:
                binary_temp, version_sum, temp_all = debug_transmission2(binary_temp[:int(lenght_sub, 2)], version_sum, temp_all)
            # print(temp_all)
            sum_all.append(do_the_function(temp_all, ID))
            binary = binary[int(lenght_sub, 2):]
        # version_sum += do_the_function(sum_all, ID)
        # sum_all = []
        # print(sum_all)
    else:
        message = ''
        # print('4',binary)
        while len(binary) >= 5:
            message += binary[1:5]
            if not int(binary[0]):
                binary = binary[5:]
                break
            binary = binary[5:]
        # print('Message',int(message, 2))
        sum_all.append(int(message, 2))
    # print('Lopulta', binary, version_sum)
    return binary, version_sum, sum_all
    
```

```python
def do_the_function(sum_all, ID):
    if ID == 0:
        return sum(sum_all)
    elif ID == 1:
        return np.product(sum_all)
    elif ID == 2:
        return min(sum_all)
    elif ID == 3:
        return max(sum_all)
    elif ID == 5:
        if sum_all[0] > sum_all[1]:
            return 1
        else:
            return 0
    elif ID == 6:
        if sum_all[0] < sum_all[1]:
            return 1
        else:
            return 0
    elif ID == 7:
        if sum_all[0] == sum_all[1]:
            return 1
        else:
            return 0
```

```python
def second_star(data):
    # print(data)
    rules = list(zip(*data[0]))
    transmission = data[1]
    All_ver_sum = 0
    for command in transmission:
        binary = ''
        # print(command)
        ver_sum = 0
        for char in command:
            binary += rules[1][rules[0].index(char)]
        # print(binary)
        jaljella, ver_sum, sum_all = debug_transmission2(binary, 0, [])
        print("lopputulos",ver_sum, sum_all)
        All_ver_sum += ver_sum
    # print('yhteensä',All_ver_sum)
    return 0
```
