```python
import time
import numpy as np
from collections import defaultdict
```

```python
def read_file():
    file = "input.txt"
    #file = "testi.txt"
    with open(file, "r") as inputfile:
        data = np.array([i.split("-") for i in inputfile.read().splitlines()])
    return data
```

```python
def run_task():
    t0 = time.time()
    data = read_file()
    #first_star(data)
    second_star(data)
    t1 = time.time()
    # print(data)
    print("aikaa meni:",round(t1-t0,4), "s")
```

```python
def find_all_paths(graph, start, end, path):
    #https://stackoverflow.com/questions/24471136/how-to-find-all-paths-between-two-graph-nodes
    path = path + [start]
    if start == end:
        return [path]
    if start not in graph:
        return []
    paths = []
    for node in graph[start]:
        #print(node.isupper())
        if node not in path or node.isupper():
            newpaths = find_all_paths(graph, node, end, path)
            for newpath in newpaths:
                paths.append(newpath)
    return paths   
```

```python
def first_star(data):
    #print(data)
    #https://www.geeksforgeeks.org/building-an-undirected-graph-and-finding-shortest-path-using-dictionaries-in-python/
    graph = defaultdict(list)
     
    # Loop to iterate over every
    # edge of the graph
    for edge in data:
        a, b = edge[0], edge[1]
         
        # Creating the graph
        # as adjacency list
        graph[a].append(b)
        graph[b].append(a)
    #print(graph)
    paths = find_all_paths(graph, 'start', 'end', [])
    print(len(paths))
    for path in paths:
        print(path)
    return 0
```
